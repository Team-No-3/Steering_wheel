## <H1>GAN: Generative Adversarial Networks<H1>

<h2>GAN (적대적 생성 신경망) 이란 ?<h2>

- 생성자(generator)와 판별자(discriminator) 두 개의 네트워크를 활용한 생성 모델
<br>
- 두 네트워크는 서로 적대적으로 경쟁하여 학습을 진행, 생성자 G는 D를 속이려고 최대한 비슷한 이미지를 만들도록 학습
- 판별자 D는 원래 이미지와 생성자 G가 만든 이미지를 잘 구분하도록 학습
- 다음의 목적 함수(objective function)를 통해 생성자는 이미지 분포를 학습
---

<h2>GAN의 목적함수<h2>

image.png

- x~pdata(x) : 실제 데이터에 대한 확률 분포에서 샘플링한 데이터
- z~pz(z) : 가우시안 분포를 사용하는 임의의 노이즈에서 샘플링한 데이터
- D(x) : 판별자, 진짜 이미지를 판별 (1에 가까우면 진짜 0에 가까우면 가짜)
- D(G(z)) : 생성자 G가 생성한 가짜 이미지 판별 ( G(z)가 1에 가까우면 진짜 0에 가까우면 가짜)
---

<H2>생성자(generator)<H2>

- 생성자는 실존하지 않지만 있을 법한 이미지를 생성할 수 있는 모델을 의미
<BR/>
- 이미지 데이터의 분포를 근사하는 모델 G를 만드는 것이 생성자의 목표
<BR/>
- 모델 G는 원래 데이터(이미지)의 분포를 근사할 수 있도록 학습
<BR/>
- 학습이 잘 되었다면 통계적으로 평균적인 특징을 가지는 데이터 (있을 법한 이미지) 를 쉽게 생성할 수 있음

image.png

<H2>판별자(discriminator)<H2>

- 현재 데이터의 샘플이 진짜 이미지인지, 아니면 Generator로부터 만들어진 것인지 <BR>구별해서 각각의 경우에 대한 확률을 평가하여 이미지가 진짜일 확률을 반환
- 판별자 D를 학습 시킬 때는 생성자 G를 고정시킨 채 실제 이미지는 높은 확률, <BR> 모조 이미지는 낮은 확률을 반환하는 방향으로 가중치를 업데이트 
---

<H2> GAN 동작 원리<H2>

image.png

<h3>
1. 실제 이미지 z가 입력<br/>
2. 판별자 D가 z를 학습<br/>
3. 생성자 G가 임의의 노이즈 데이터를 사용하여 모조 이미지 G(z)를 생성<br>
4. 생성 된 G(z)를 다시 판별자 D에 입력을 하면 G(z)가 실제 이미지일 확률을 반환
<h3>

## <H1>AnoGAN: Unsupervised Anomaly Detection with GAN<H1>

<h2>AnoGAN 이전의 Anomaly Detection<h2>

- 기존 Anomaly Detection은 Supervised Learning으로 접근함
- 산업현장에서는 Normal 데이터는 많지만 Abnormal 데이터는 부족한 경우 대부분이므로 데이터를 확보하는데 문제가 발생함
- 보통의 모델은 자동 감지를 위해 알려진 마커의 주석이 달린 예제를 사용하여 많은 양의 데이터를 기반으로 하기 때문에, 많은 시간과 비용이 발생함
- 주석처리 작업이 이루어진 데이터에 대해서만 탐지 가능하다는 단점이 발생함
---

<h2>ANOGAN 등장 계기<h2>


- 의료현장, 산업현장 모두 공통적으로 Normal 데이터는 많지만 Abnormal 데이터는 부족
- 지금까지 Anomaly Detection을 하기 위해 Supervised Learning 방법으로 진행, 이는 Anotation된 데이터가 많이 필요하다는 단점이 발생
- Unsupervised Learning 방법 중 GAN을 이용한 Anomaly Detection을 진행
- Supervised Learning으로 문제를 해결하려면 우선 Data Imbalance 문제를 해결해야함. 이런 이유로 Unsupervised Learning 방법으로 접근
---

<h2>AnoGAN 이란?<h2>

- anoGAN은 정상적인 변동성의 다양함(정상 데이터의 분포)를 학습한 DCGAN과 이미지 공간에서 잠재공간으로의 매핑방식을 기반으로 Anomaly Score를 계산하고 새로운 데이터에 적용했을때 모델은 Anomaly와 Anomaly Score를 계산
- DCGAN에서 이루어진 이미지 분포의 학습을 통해 학습된 이미지와 다른 변칙적인 영상(Anomaly)을 정확하게 식별
